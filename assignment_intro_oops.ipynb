# 1. Digital Bookstore Inventory

**Objective:**  
Create a Python class to model a book in a digital bookstore inventory system.

---

### **Requirements**
- Create a class named `Book`.
- The constructor `__init__` must accept three parameters:
  - `title`: A string representing the bookâ€™s title.
  - `author`: A string representing the bookâ€™s author.
  - `price`: A float representing the bookâ€™s price.
- Store these parameters as instance attributes.
- Implement a method `get_details()` that returns a formatted string in the format:  
  `"[Title]" by [Author], Price: $[Price]"`
  - The price must be formatted to exactly two decimal places.
  - The title must be enclosed in double quotes.
- Ensure the class handles the sample usage correctly.

---

### **Constraints**
- `title` and `author` will always be non-empty strings.
- `price` will always be a positive number.

---

### **Sample Usage**
```python

print(book1.get_details())

---

### **Expected Output
"The Hitchhiker's Guide to the Galaxy" by Douglas Adams, Price: $12.50


Task

Write the complete Book class implementation.

````python
class Book:
    def __init__(self,title,author,price):
        self.title=title
        self.author=author
        self.price=price

    def get_details(self):
        str=f'"{self.title}" by {self.author}, Price: ${self.price:.2f}'
        return str
    
book1=Book("The Hitchhiker's Guide to the Galaxy", "Douglas Adams", 12.5)
print(book1.get_details())
````

## 2. Employee Hierarchy at a Tech Startup
 
**Objective:**  
Model an employee hierarchy using inheritance to represent employees and managers at a tech startup.
 
---
 
### Requirements:
- Create a base class named `Employee` with:
  - An `__init__` method that accepts:
    - `name`: A string representing the employeeâ€™s name.
    - `salary`: An integer representing the employeeâ€™s salary.
  - Store these as instance attributes.
  - A method `display_role()` that returns the string `"Employee"`.
 
- Create a derived class named `Manager` that inherits from `Employee` with:
  - An `__init__` method that accepts:
    - `name`: A string.
    - `salary`: An integer.
    - `team_size`: An integer representing the number of direct reports.
  - Call the parent classâ€™s `__init__` method to initialize `name` and `salary`.
  - Store `team_size` as an instance attribute.
  - Override the `display_role()` method to return the string `"Manager"`.
 
- Ensure the classes handle the sample usage correctly.
 
---
 
### Constraints:
- `name` will be a non-empty string.
- `salary` and `team_size` will be positive integers.
 
---
 
### Sample Usage:
```python
emp = Employee("Alice", 80000)
mgr = Manager("Bob", 120000, 5)
 
print(f"{emp.name} is an {emp.display_role()} with a salary of ${emp.salary}.")
print(f"{mgr.name} is a {mgr.display_role()} with a salary of ${mgr.salary} and manages a team of {mgr.team_size}.")

---

### Expected Output:

Alice is an Employee with a salary of $80000.
Bob is a Manager with a salary of $120000 and manages a team of 5.

---
### Task:

Write complete Employee and Manager class implementations.

````python
class Employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary

    def display_role(self):
        return "Employee"
    
class Manager(Employee):
    def __init__(self, name, salary,team_size):
        super().__init__(name, salary)
        self.team_size=team_size

    def display_role(self):
        return "Manager"
    
emp = Employee("Alice", 80000)
mgr = Manager("Bob", 120000, 5)
 
print(f"{emp.name} is an {emp.display_role()} with a salary of ${emp.salary}.")
print(f"{mgr.name} is a {mgr.display_role()} with a salary of ${mgr.salary} and manages a team of {mgr.team_size}.")
````

---
 
### ðŸ§© Requirements
 
- Create a base class named **Animal** with:
  - A method `make_sound()` that returns `"Some generic animal sound"`
 
- Create two derived classes, **Lion** and **Monkey**, that inherit from `Animal`:
  - In the **Lion** class, override `make_sound()` to return `"Roar!"`
  - In the **Monkey** class, override `make_sound()` to return `"Ooh ooh aah aah!"`
 
- Write a standalone function `trigger_sounds(animals)` that:
  - Accepts a list of `Animal` objects.
  - Calls the `make_sound()` method on each object and prints the result on a new line.
 
- Ensure the classes and function handle the sample usage correctly.
 
---
 
### ðŸ§ª Sample Usage
 
```python
leo = Lion()
george = Monkey()
generic_animal = Animal()

### Expected output

Roar!
Ooh ooh aah aah!

---

### Task

write complete Animal, Lion, Monkey class implementations, slong with trigger_sounds function.


````python
class Animal:
    def make_sound(self):
        return "Some generic animal sound"
    
class Lion(Animal):
    def make_sound(self):
        return "Roar!"
    
class Monkey(Animal):
    def make_sound(self):
        return "Ooh ooh aah aah!"
    
def trigger_sounds(animals):
    for animal in animals:
        print(animal.make_sound())

leo = Lion()
george = Monkey()
generic_animal = Animal()
 
animal_list = [leo, george, generic_animal]
trigger_sounds(animal_list)
````

### ðŸª Design a Simple Vending Machine
 
**Objective:**  
Design a set of Python classes to model a vending machine that sells products, accepts money, gives change, and tracks inventory.
 
---
 
### ðŸ§© Requirements
 
#### Create a `Product` class with:
- **Attributes:**
  - `name`: A string representing the product name.  
  - `price`: An integer representing the price in cents.  
  - `quantity`: An integer representing the available quantity.

---
 
#### Create a `VendingMachine` class with:
- **Attributes:**
  - `inventory`: A dictionary where keys are product names (strings) and values are `Product` objects.  
  - `current_balance`: An integer representing the money inserted (in cents), defaulting to 0.  
 
- **Methods:**
  - `load_inventory(products)`:  
    Accepts a dictionary with product names as keys and dictionaries containing price and quantity as values.  
    Populates `inventory` with `Product` objects.  
    Example input:  
    ```python
    {"Soda": {"price": 125, "quantity": 5}, "Chips": {"price": 100, "quantity": 3}}
    ```
  
  - `display_products()`:  
    Displays all available products with their name, price, and quantity.
 
  - `insert_money(amount)`:  
    Adds `amount` to `current_balance`.
 
  - `purchase(product_name)`:  
    - Checks if the product exists. If not, return `"Error: Product not found."`
    - Checks if the product is in stock. If not, return `"Error: [product_name] is out of stock."`
    - Checks if the balance is enough. If not, return `"Error: Insufficient funds."`
    - Deducts price from balance, reduces product quantity by 1, and returns:  
      `"Purchased [product_name]. Your change is [change] cents."`
 
---
 
### âš™ï¸ Constraints
 
- All monetary values are integers in cents (no floating point numbers).  
- Sample data uses small integer values (in cents for prices and quantities).

### ðŸ§ª Sample Input
 
```python

    "Soda": {"price": 150, "quantity": 5},
    "Chips": {"price": 100, "quantity": 10},

### Task

Write complete Product and VendingMachine class implementation

````python
class Product:
    
    def __init__(self,name:str,price:float,quantity:int):
        self.name=name
        self.price=price
        self.quantity=quantity
    
    def decrease_quantity(self):
        if self.quantity>0:
            self.quantity=self.quantity-1

class VendingMachine:
    inventory={}
    current_balance=0
    
    def __init__(self,inventory:dict):
        super().__init__()
        self.inventory=inventory

    def display_products(self):
        for name,details in self.inventory.items():
                price=details['price']
                quantity=details['quantity']
                if quantity>0:
                    print(f"Product: {name}, Price: {price} cents, Quantity: {quantity}")
    
    def insert_coin(self,amount):
        self.current_balance+=amount

    def purchase_product(self,product_name):
        flag=0
        for name,details in self.inventory.items():
            
            if name==product_name:
                flag+=1
                price=details['price']
                quantity=details['quantity']
                if quantity==0:
                    return f"Error: {product_name} is out of stock."
                elif self.current_balance<price:
                    return f"Error: Insufficient funds."
                else:
                    quantity-=1
                    change=self.current_balance-price
                    self.current_balance=0
                    return f"Purchased {product_name}. Your change is {change} cents."
        if flag==0:    
            return f"Error: Product '{product_name}' not found."
                
            

            

initial_inventory = {
    "Soda": {"price": 150, "quantity": 5},
    "Chips": {"price": 100, "quantity": 10},
    "Candy": {"price": 75, "quantity": 0}
}
vm = VendingMachine(initial_inventory)
print("--- Available Products ---")
vm.display_products()
print("\n--- Successful Purchase ---")
vm.insert_coin(100)
vm.insert_coin(50)
print(vm.purchase_product("Soda"))
print(f"Current balance: {vm.current_balance}")
print("\n--- Insufficient Funds ---")
vm.insert_coin(50)
print(vm.purchase_product("Chips"))
print(f"Current balance: {vm.current_balance}")
print("\n--- Out of Stock ---")
vm.insert_coin(100)
print(vm.purchase_product("Candy"))
````
